```python
import numpy as np
import pandas as pd
```

```python
L = pd.read_csv("data_worldbank.csv")
```

```python
L.head()
```

```python
L.duplicated().sum()
```

```python
L.drop_duplicates(inplace=True)
```

```python
L.info()
```

```python
L.isnull().sum()
```

```python
L.dropna(subset=['Board Approval Date', 'Effective Date (Most Recent)'], inplace=True)
```

```python
L.drop(columns=["Country Code", "Borrower's Obligation (US$)"], inplace=True)
```

```python
new_columns = {
    'Original Principal Amount (US$)': 'Original Principal Amount',
    'Undisbursed Amount (US$)': 'Undisbursed Amount',
    'Disbursed Amount (US$)': 'Disbursed Amount',
    'Repaid to IDA (US$)': 'Repaid to IDA',
    'Due to IDA (US$)': 'Due to IDA',
    "Cancelled Amount (US$)": "Cancelled Amount",
    'Effective Date (Most Recent)': 'Effective Date'
}

L.rename(columns=new_columns, inplace=True)
```

```python
date_columns = ['Board Approval Date', 'Effective Date', 'End of Period']
L[date_columns] = L[date_columns].apply(pd.to_datetime, format='%m/%d/%Y %I:%M:%S %p')
```

```python
L.to_csv('cleaned_dataset.csv', index=False)
```
